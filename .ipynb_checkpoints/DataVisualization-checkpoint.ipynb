```python
# Libraries/packages
import os
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# directory path for ploting, reading, writing results

base_path=(r'UASB_HRAP_Modelling')
data_path=os.path.join(base_path,'Data','Dynamic Simulation')
plot_path=os.path.join(base_path,'Plot', 'Correlation')
results_dir  = os.path.join(base_path, 'Results', 'Soft-Sensing','FFNN')

# read input and output: Influent and effluent
influent=pd.read_csv(os.path.join(data_path,'DynamicInfluent-hour.csv'))
effluent=pd.read_csv(os.path.join(data_path,'HRAP-Feb-Group-New.csv'))

uasb=pd.read_csv(os.path.join(data_path,'UASB-hour.csv'))
print(influent)

influent=influent.loc[:, (influent != 0).any(axis=0)] # remove columns with zeros
print(influent.columns)
effluent=effluent.loc[:, (effluent != 0).any(axis=0)]
print(effluent.columns)
uasb=uasb.loc[:, (uasb != 0).any(axis=0)]
print(uasb)


# we can iterate through multiple dataFrames for ploting etc.
data_files = {
        'influent': pd.DataFrame(influent),
        'uasb': pd.DataFrame(uasb),
        'effluent':pd.DataFrame(effluent)
}
for key,val in data_files.items():
    #print(val)

# plot correaltions
    fig, ax = plt.subplots(figsize=(25, 25))
    sns.heatmap(val.corr().round(2), annot=True, ax=ax,cmap="YlGnBu")#heatmap

    bottom, top = ax.get_ylim()
    ax.set_ylim(bottom + 0.5, top - 0.5)
    ax.set_title(key+ " VS " +key+" Correlations")
#plt.savefig(os.path.join(plot_path,"Influent-Cross-Corr"))
    plt.show()
```
