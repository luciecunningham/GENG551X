```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
features = ["Absorbance", "Cell Count", "COD", "DO", "FIA", "Influent", "pH", "Temperature", "TSS&VSS", "Turbidity"]
files = []
for feature in features:
    file = feature + ".csv"
    data_path = os.path.join("data", file)
    files.append(data_path)
```

```python
df0 = pd.read_csv(files[0])
#what is overflow? Answer: Overflow is excess and not counted so removed
#Remove "Unnamed: 5" as it is a empty column
#Remove days since start as it adds no value (with date being a more useful measurement)
to_remove = ["Unnamed: 5", "Days Since Start", "RW2 Overflow", "RW3/1 Overflow"]
df0 = df0.drop(columns=to_remove)
df0.info()
```

```python
df1 = pd.read_csv(files[1])
to_remove = ["Unnamed: 5", "Days Since Start", "Unnamed: 7", "RW2 Overflow", "RW3 Overflow"]
df1 = df1.drop(columns=to_remove)
df1.info()
```
