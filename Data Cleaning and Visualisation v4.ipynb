```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns

import json
import os
import datetime as dt

from pandas.plotting import lag_plot
import matplotlib.dates as mdates

import xlrd
```

```python
total_data_file = os.path.join("data", "PROBE-Lab-WS- Merged.csv")
total_df = pd.read_csv(total_data_file)
```

```python
total_df = total_df[~(total_df["D.O. (mg/L)"] < 2.4)]
```

```python
total_df = total_df.rename(columns={'Temperature (C)': 'Temperature (°C)'})
```

```python
total_df = total_df[['pH', 'Temperature (°C)', 'D.O. (mg/L)', 'Cell Count (million per mL)','NO3-N (mg/L)', 'NH4-N (mg/L)', 'ORP (MV)', 'Conductivity (mS/cm)', 'Turbidity (NTU)']]
```

```python
total_df = total_df[~(total_df["Turbidity (NTU)"] > 200)]
total_df = total_df[~(total_df["Turbidity (NTU)"] < 20)]
```

```python

plt.scatter(total_df["Cell Count (million per mL)"], total_df["Turbidity (NTU)"])
plt.xlabel("Cell Count (million per mL)")
plt.ylabel("Turbidity (NTU)")
plt.title("Correlation Between Cell Count and Turbidity")
#plt.savefig(feature.split(" ", 2)[0] + " v Cell Count.png")
plt.show
```

```python
output_datafile = os.path.join("data", "final_df.csv")
total_df.to_csv(output_datafile, index=False)
```
